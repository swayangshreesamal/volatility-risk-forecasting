```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/My Drive/volatility-risk-forecasting/DATA/PROCESSED/nifty50_processed.csv", index_col=0, parse_dates=True)
df
```

```python
TRADING_DAYS=252
df['vol_annualized'] =df['log_return'].std()*np.sqrt(TRADING_DAYS)
df
```

```python
df['vol_14d_annualized']=df['vol_14d'] * np.sqrt(TRADING_DAYS)
df['vol_30d_annualized']=df['vol_30d'] * np.sqrt(TRADING_DAYS)
df
```

```python
df[['vol_14d_annualized', 'vol_30d_annualized']].tail()
df
```

```python
df['return_14d']=df['log_return'].rolling(window=14).mean() *14
df['return_14d']
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['vol_14d_annualized'],df['return_14d'],alpha=0.4)
plt.xlabel("Annualized Volatility (14d)")
plt.ylabel("14-Day Return")
plt.title("Risk vs Return Tradeoff (NIFTY 50)")
plt.show()
```
